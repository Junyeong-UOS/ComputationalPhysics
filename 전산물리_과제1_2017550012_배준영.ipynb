### 과제1. 3번

```python
import numpy as np
import matplotlib.pyplot as plt
ey = np.array([0,1])
for alpha in np.arange(-2,2,0.1):
    A = np.array([[1,alpha],
                  [alpha,1]])
    v = np.dot(A,ey)
    plt.arrow(0,0,v[0],v[1],head_width=0.05,color='r')

plt.arrow(0,0,ey[0],ey[1],head_width=0.05,color='k')
plt.text(-0.3,1.2,r'$e_y$ = (0,1)',fontdict={'size': 13})
plt.axhline(y=1,linewidth=3,color='b')
plt.xlim(-2,2)
plt.ylim(0,2)
plt.show()
```

### 과제1. 4번

```python
import numpy as np
ex = np.array([1,0])
ey = np.array([0,1])
u = ex + ey   
v = ex - ey
#A = np.array([[a11,a12],[a21,a22]]) (2x2 matrix)
# Au = Aex + Aey / Av = Aex - Aey
# a11  + a12 = 3 / a21 + a22 = 2
# a11 -a12 = -1  / a21 - a22 = -2

Sol = np.array([[1,1], 
                 [1,-1]])

Atop = ([3,-1])
Abot = ([2,-2])

AnsAtop = np.linalg.solve(Sol,Atop)
AnsAbot = np.linalg.solve(Sol,Abot)

A = np.vstack([AnsAtop, AnsAbot])

print(np.dot(A,u))
print(np.dot(A,v))
print(A.shape)
print('A =''\n',A)
```

### 과제1. 5번

```python
import numpy as np
#토끼=x,4 / 닭=y,2
# 4*x + 2*y = 92
# x + y = 40

A = np.array([[4,2],
               [1,1]])
b = np.array([92,40])

x = np.linalg.solve(A,b)
print('우리 안에는 ', end='')
print('토끼',int(x[0]),'마리와 ',end='')
print('닭',int(x[1]),'마리가 있습니다.')
```
